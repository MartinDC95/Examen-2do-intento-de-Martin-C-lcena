```python
import pandas as pd

# 'WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
# Limpieza de datos
df.dropna(inplace=True)  # Eliminar filas con valores NaN
```

```python
# Verificar duplicados
duplicados = df.duplicated().sum()
print(f'Número de duplicados: {duplicados}')

# Eliminar duplicados
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ejemplo de visualización
plt.figure(figsize=(10, 6))
sns.histplot(df['MonthlyCharges'], bins=30)
plt.title('Pagos mensuales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['MonthlyCharges'])
plt.title('Pagos mensuales')
plt.show()
# Interpretación: Aquí puedes mencionar si hay valores atípicos, mediana, etc.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ejemplo de visualización
plt.figure(figsize=(10, 6))
sns.histplot(df['PaymentMethod'], bins=30)
plt.title('Metodo de Pago')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame

# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Valores en cantidad')
plt.show()
# Interpretación: Comenta sobre las correlaciones fuertes o débiles observadas.
```
